```python
!pip install aksharamukha
from aksharamukha import transliterate
import time


def transliterate_file(
    infile,
    outfile
):

    start = time.time()

    count = 0

    with open(
        infile,
        encoding="utf-8"
    ) as fin, open(
        outfile,
        "w",
        encoding="utf-8"
    ) as fout:

        for line in fin:

            txt = line.strip()

            if not txt:
                continue

            iso = transliterate.process(
                "Bengali",
                "ISO",
                txt
            )

            fout.write(
                iso + "\n"
            )

            count += 1

            if count % 10000 == 0:

                elapsed = (
                    time.time()
                    - start
                )

                print(
                    f"{count:,} docs | "
                    f"{count/elapsed:.2f} docs/sec"
                )

    print(
        "Completed:",
        count
    )
```

```python
from collections import Counter


def corpus_stats(fname):

    chars = 0
    words = 0

    vocab = Counter()

    with open(
        fname,
        encoding="utf-8"
    ) as f:

        for line in f:

            chars += len(line)

            ws = line.split()

            words += len(ws)

            vocab.update(ws)

    return {

        "chars": chars,

        "words": words,

        "vocab": len(vocab),

        "avg_word_len":
        sum(
            len(w)
            for w in vocab
        ) / len(vocab)

    }


bn = corpus_stats(
    "train_social_bn.txt"
)

iso = corpus_stats(
    "train_social_iso.txt"
)

print(
    "Bengali:",
    bn
)

print(
    "ISO:",
    iso
)

print()

print(
    "Expansion Factor:",
    iso["chars"]
    /
    bn["chars"]
)

print(
    "Vocabulary Ratio:",
    iso["vocab"]
    /
    bn["vocab"]
)
```

```python
from tokenizers import Tokenizer
import numpy as np

def evaluate_bpe(
    tokenizer_file,
    test_file,
    vocab_size
):

    tokenizer = Tokenizer.from_file(
        tokenizer_file
    )

    total_words = 0
    total_tokens = 0
    total_chars = 0

    fragmented = 0
    oov = 0

    token_counts = []

    with open(
        test_file,
        encoding="utf-8"
    ) as f:

        for line in f:

            words = line.strip().split()

            total_words += len(words)

            sent_tokens = 0

            for word in words:

                tokens = tokenizer.encode(
                    word
                ).tokens

                total_tokens += len(tokens)

                sent_tokens += len(tokens)

                total_chars += len(word)

                if len(tokens) > 1:
                    fragmented += 1

                if "[UNK]" in tokens:
                    oov += 1

            token_counts.append(
                sent_tokens
            )

    fertility = (
        total_tokens /
        total_words
    )

    cpt = (
        total_chars /
        total_tokens
    )

    compression = cpt

    avg_tokens = np.mean(
        token_counts
    )

    wfr = (
        fragmented /
        total_words
    )

    variance = np.var(
        token_counts
    )

    return {
        "vocab": vocab_size,
        "oov": oov / total_words,
        "fertility": fertility,
        "cpt": cpt,
        "compression": compression,
        "avg_tokens": avg_tokens,
        "wfr": wfr,
        "variance": variance
    }
```

```python
def evaluate_unigram(
    tokenizer_file,
    test_file,
    vocab_size
):

    from tokenizers import Tokenizer
    import numpy as np

    tokenizer = Tokenizer.from_file(
        tokenizer_file
    )

    total_words = 0
    total_tokens = 0
    total_chars = 0

    fragmented = 0
    token_counts = []

    with open(
        test_file,
        encoding="utf-8"
    ) as f:

        for line in f:

            words = line.strip().split()

            total_words += len(words)

            sent_tokens = 0

            for word in words:

                try:
                    tokens = tokenizer.encode(
                        word
                    ).tokens

                except:
                    continue

                total_tokens += len(tokens)
                sent_tokens += len(tokens)
                total_chars += len(word)

                if len(tokens) > 1:
                    fragmented += 1

            token_counts.append(
                sent_tokens
            )

    fertility = total_tokens / total_words
    cpt = total_chars / total_tokens
    compression = cpt

    return {
        "vocab": vocab_size,
        "oov": 0.0,
        "fertility": fertility,
        "cpt": cpt,
        "compression": compression,
        "avg_tokens": np.mean(token_counts),
        "wfr": fragmented / total_words,
        "variance": np.var(token_counts)
    }
```

```python
VOCABS = [
    5000,
    10000,
    15000,
    20000,
    25000,
    30000,
    35000,
    40000,
    45000,
    50000
]

uni_social_bn_results = []

for vocab_size in VOCABS:

    res = evaluate_unigram(
        f"uni_social_bn_{vocab_size}.json",
        "test_social_bn.txt",
        vocab_size
    )

    print(res)

    uni_social_bn_results.append(
        res
    )

print("\nFinished.")
```

## Social Media Dataset Summary

The Social Media corpus was constructed from the Bangla Social Media Abuse Dataset containing 44,001 user-generated comments collected from online social platforms. The corpus represents informal, user-generated text and includes a wide variety of linguistic phenomena such as colloquial expressions, spelling variations, abbreviations, emotive language, code-mixing, and non-standard orthographic forms that are commonly observed in social media communication.

After preprocessing, the corpus contained 731,153 words and 4,209,905 characters with a vocabulary size of 96,307 unique word types. The Type-Token Ratio (TTR) of 0.1317 was the highest among all evaluated domains, indicating substantial lexical diversity. The average word length was 7.60 characters, reflecting the presence of both standard lexical items and elongated or expressive social-media spellings.

The corpus was divided into training and testing subsets using a 90:10 split, resulting in 39,701 training documents and 4,412 testing documents. The training corpus occupied 9.42 MB, while the testing corpus occupied 1.06 MB. To facilitate script-based comparison, the Bengali corpus was transliterated into ISO 15919 using Aksharamukha. The transliterated corpus exhibited an expansion factor of 1.1693, indicating an increase in character count due to the explicit representation of phonological distinctions in the Romanized script. The vocabulary ratio between the ISO and Bengali versions was 0.9723, suggesting minimal vocabulary loss during transliteration.

A round-trip transliteration evaluation yielded an exact-match accuracy of 74%, which was notably lower than that observed for the news and literature domains. This reduction can be attributed to the inherently noisy nature of social media text, including spelling inconsistencies, informal writing styles, punctuation variations, emojis, and other non-standard textual elements. Despite this noise, the corpus provides an important benchmark for evaluating tokenizer robustness under realistic user-generated language conditions.

Experimental results demonstrated that Byte Pair Encoding (BPE) consistently achieved the best segmentation performance across both Bengali and ISO scripts, followed closely by WordPiece, while Unigram produced comparatively higher fertility and fragmentation rates. Interestingly, despite exhibiting the highest lexical diversity among all domains, the social media corpus achieved the lowest fertility values under BPE tokenization, suggesting that the repetitive nature of user-generated comments enables efficient subword learning despite substantial vocabulary variation.
